```python
import numpy as np
import matplotlib.pyplot as plt

# Generate time series
t = np.linspace(0, 2*np.pi, 100)  # One full AC cycle

# Three-phase AC currents (sinusoids with 120-degree phase shifts)
i_a = np.sin(t)
i_b = np.sin(t - 2*np.pi/3)
i_c = np.sin(t + 2*np.pi/3)

# Plot the three-phase currents
plt.figure(figsize=(10,5))
plt.plot(t, i_a, label="i_a")
plt.plot(t, i_b, label="i_b")
plt.plot(t, i_c, label="i_c")
plt.xlabel("Time")
plt.ylabel("Current")
plt.title("Three-phase currents")
plt.legend()
plt.grid()
plt.show()


# Clarke Transformation Equations
i_alpha = (2/3) * (i_a - 0.5*i_b - 0.5*i_c)
i_beta = (2/3) * ((np.sqrt(3)/2) * (i_b - i_c))

# Plot Clarke Transform results
plt.figure(figsize=(10,5))
plt.plot(t, i_alpha, label="i_alpha")
plt.plot(t, i_beta, label="i_beta")
plt.xlabel("Time")
plt.ylabel("Current")
plt.title("Clarke Transformation Results")
plt.legend()
plt.grid()
plt.show()


# Define rotor speed (can be varied)
omega = 1  # Rotational speed (electrical rad/sec)
theta = omega * t  # Theta must change over time!

# Park Transformation Equations
i_d = i_alpha * np.cos(theta) + i_beta * np.sin(theta)
i_q = -i_alpha * np.sin(theta) + i_beta * np.cos(theta)

# Plot Park Transform results
plt.figure(figsize=(10,5))
plt.plot(t, i_d, label="i_d (Direct-axis Current)")
plt.plot(t, i_q, label="i_q (Quadrature-axis Current)")
plt.xlabel("Time")
plt.ylabel("Current")
plt.title("Park Transformation Results")
plt.legend()
plt.grid()
plt.show()


```
